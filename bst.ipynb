# BINARY SEARCH TREE

```python
class node:
    def __init__(self,v):
        self.left = None
        self.info = v
        self.right = None
class Bst:
    def __init__(self):
        self.root = None
    def creating(self,v):
        n = node(v)
        if self.root == None:
            self.root = n
            return
        temp = self.root
        while temp  != None:
            if n.info <= temp.info :
                par = temp
                temp = temp.left
            elif n.info > temp.info:
                par = temp
                temp = temp.right
        if n.info <= par.info:
            par.left = n
        else:
            par.right = n
    def inorder(self,node):
        if node != None:
            self.inorder(node.left)
            print(node.info)
            self.inorder(node.right)
    def delete(self,item):
        temp = self.root
        while temp != None and temp.info != item:
            if item <= temp.info:
                par = temp
                temp = temp.left
            elif item > temp.info:
                par = temp
                temp = temp.right
        def del_no_child(self,temp,par):
            if temp.info <= par.info:
                par.left = None
                del temp
            else:
                par.right = None
                del temp


        def del_single_left_child(self,temp,par):
            if temp.info <= par.info:
                par.left = temp.left
                del temp
            else:
                par.right = temp.left
                del temp

        def del_single_right_child(self,temp,par):
            if temp.info <= par.info:
                par.left = temp.right
                del temp
            else:
                par.right = temp.right
                del temp
        def two_child(self,temp,par):
            succ = temp.right
            succ_par = temp
            while succ.left != None:
                succ_par = succ
                succ = succ.left
            delete_Node = succ.info
            
            if succ.right == None:
                del_no_child(self,succ,succ_par)
            else:
                del_single_right_child(self,succ,succ_par)
            temp.info = delete_Node
        if temp == None:
            print("Item not Found ")
            return
        else:
            if temp.left == None and temp.right == None:
                del_no_child(self,temp,par)
            elif temp.left != None and temp.right ==None:
                del_single_left_child(self,temp,par)
            elif temp.left == None and temp.right != None:
                del_single_right_child(self,temp,par)
            else:
                two_child(self,temp,par)
        del temp
    def height(self,st):
        if st == None:
            return 0
        else:
            return max(self.height(st.left),self.height(st.right))+1
```

```python
st = Bst()
st.creating(89)
st.creating(40)
st.creating(90)
st.inorder(st.root)
st.creating(80)
st.creating(160)
st.creating(60)
st.creating(30)
print('\n')
st.inorder(st.root)
st.delete(60)
print('Item deleted sucssecfully \n')
st.inorder(st.root)
print(f'Height of the Binary Search Tree is {st.height(st.root)}')
```
