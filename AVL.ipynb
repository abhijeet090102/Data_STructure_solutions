```python
class node:
    def __init__(self,v):
        self.left = None
        self.info = v
        self.right = None
class AVL:
    def __init__(self):
        self.root = None
    def creating(self,v):
        n = node(v)
        if self.root == None:
            self.root = n
            return
        temp = self.root
        while temp  != None:
            if n.info <= temp.info :
                par = temp
                temp = temp.left
            elif n.info > temp.info:
                par = temp
                temp = temp.right
        if n.info <= par.info:
            par.left = n
        else:
            par.right = n
        if self.balanceFactor(self.root) not in [-1,0 ,1]:
            print(' Tree is unbalance ')
            
        temp = self.root
        if self.balanceFactor(temp) in [0,-1,1]:
            return
        pivot = None
        pivo_par = None
        while self.balanceFactor(temp) not in [0,1,-1]:
            if pivot is not None:
                pivo_par = pivot
            if v < temp.info:
                pivot = temp
                temp = temp.left
            else:
                pivot = temp
                temp = temp.right
        if temp.info < pivot.info and v<temp.info:
            print(pivot.info)
            print(temp.info)
            print("LL rotation can happend ")
            self.llRotation(pivot,pivo_par )

        elif temp.info > pivot.info and v > temp.info:
            print("RR rotation can happend ")
            self.rrRotation(pivot,pivo_par)
        elif temp.info > pivot.info and v < temp.info :
            print("RL rotation can happen ")
            self.rlRotation(pivot,pivo_par)
        elif temp.info < pivot.info and v > temp.info:
            print('LR Rotaiton can happen ')
            self.lrRotaiton(pivot,pivo_par)
            
    def llRotation(self,piv,pivo_par):
        temp = piv.left
        piv.left = temp.right
        temp.right = piv
        if pivo_par == None:
            self.root = temp
        else:
            if pivo_par.left == piv:
                pivo_par.left = temp
            else:
                pivo_par.right = temp
            
    def rrRotation(self,piv,pivo_par):
        temp = piv.right
        piv.right= temp.left
        temp.left = piv
        if pivo_par == None:
            self.root = temp
        else:
            if pivo_par.right == piv:
                pivo_par.right = temp
            else:
                pivo_par.left = temp
        
    def rlRotation(self,piv,pivo_par):
        '''temp = piv.right
        a = temp.left
        piv.left = a.left 
        temp.right = a
        a.left = temp'''
        self.llRotation(piv.right,piv)
        self.rrRotation(piv,pivo_par)
    def lrRotaiton(self,piv,pivo_par):
        self.rrRotation(piv.left,pivo_par)
        self.llRotation(piv,pivo_par)
        
    def height(self,rt):
        if rt == None:
            return 0
        else:
            return max(self.height(rt.left),self.height(rt.right))+1
    def balanceFactor (self,n):
        return (self.height(n.left)-self.height(n.right))
        
            
    def inorder(self,node):
        if node != None:
            self.inorder(node.left)
            print(node.info)
            print(self.balanceFactor(node))
            self.inorder(node.right)
            
```
